### Example1-with curvefault_velocity_model

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from forward import Wave_Forward
matplotlib.use('TkAgg')
```

```python
v = np.load('data/curvefault_velocity_model_sample.npy')
model = Wave_Forward(velocity_model=v,
                     x_grid_nums=70,
                     z_grid_nums=70,
                     dx=10,
                     bgnums=200,
                     nt=1000,
                     dt=0.001,
                     freq=15,
                     s_loc=(350,0),
                     r_loc=[(i,0)for i in range(0,700,10)],
                     save_seis='seismic.npy',
                     save_wavefield='wavefield.npy')
# 正演
model.forward()
```

```python
%matplotlib inline
# 正演结果
plt.figure(figsize=(2,2))
plt.ion()
plt.imshow(model.seis,aspect='auto',vmax=0.6,vmin=-0.2,cmap='gray')
plt.show()
```

```python
%matplotlib inline
# 真实值
real = np.load('./data/curvefault_seismic_wave_model_sample.npy')
plt.figure(figsize=(4,4))
plt.ion()
plt.imshow(real[2],aspect='auto',vmax=0.6,vmin=-0.2,cmap='gray')
plt.show()
```

```python
%matplotlib
# 波场图
plt.ion()
for wave in model.wavefield:
    plt.clf()
    plt.imshow(wave,cmap='rainbow',vmin=-2.5,vmax=2.5)
    plt.colorbar()
    plt.pause(0.05)
plt.close()
```
