```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

#### create velocity model

```python
vel = np.ones((70,70,70))*2000
```

FD

```python
from forward3d import Acoustic_FD
from utils import *
```

```python
# 参数配置
tm = time_model(dt=0.001,T=1)
sm = spatial_model(vel,10)
so = source(time_model=tm,location=(35,35,0),freq=20)
x = 35*np.ones(70).astype(int)
y = np.arange(70).astype(int)
z = 69*np.ones(70).astype(int)
receiver_loc = np.c_[x,y,z]
rec = receiver(location=receiver_loc)
```

```python
fd = Acoustic_FD(time_model=tm,
                 spatial_model=sm,
                 source=so,
                 receiver=rec)
seis = fd.forward()
```

```python
plt.imshow(seis,vmax=0.1,vmin=-0.1,aspect=70/1000,cmap='gray')
plt.colorbar()
plt.show()
```
